# Data Visualization

```python
amazon.columns
```

```python
ax=sns.countplot(x='Gender',data=amazon)
for bars in ax.containers:
    ax.bar_label(bars)
```

Based on the graph above, it appears that male members tend to subscribe to Amazon Prime more frequently than female members

```python
subscription=sns.countplot(x='Subscription Plan',data=amazon)
for bar in subscription.containers:
    subscription.bar_label(bar)
```

The graph above indicates that a higher proportion of people opt for the yearly plan compared to the monthly plan

```python
Devices=amazon.groupby('Devices Used').size().reset_index().sort_values(by=0,ascending=False)
Devices=sns.barplot(y='Devices Used',x=0,data=Devices)
for bar in Devices.containers:
    Devices.bar_label(bar)
```

The graphs above indicate that a majority of people prefer using smartphones for streaming 

```python
rating=amazon.groupby('Feedback/Ratings').size().reset_index().sort_values(by=0,ascending=False)
rating=sns.barplot(x='Feedback/Ratings',y=0,data=rating)
for bar in rating.containers:
    rating.bar_label(bar)
```

According to the graph, the majority of people have rated Amazon Prime content at 4.3 stars.

```python

customer_suport=sns.countplot(x='Customer Support Interactions',data=amazon)
for bar in customer_suport.containers:
    customer_suport.bar_label(bar)
```

The graph shows that 1 star rating received the most votes, with a total of 241 votes

```python
genres=amazon.groupby('Favorite Genres').size().reset_index()
genres=sns.countplot(y='Favorite Genres',data=amazon)
for bar in genres.containers:
    genres.bar_label(bar)
```

The graph above indicates that people have a preference for watching horror content, followed by action and romance genres.
